```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


dataset_url = ('https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv')
df = pd.read_csv(dataset_url, parse_dates=['Date'], index_col='Date')
pd.set_option('display.max.columns', None)
print(df.head())
```

1. Line Plot: instead of ‘MSFT’ plot for ‘AAPL’

```python
df.plot(y='AAPL', figsize=(9,6))
```

2. Multiline Plot: change the color of the three lines to blue, magenta and yellow.

```python
df.plot.line(y=['FB','AAPL','MSFT'],color='bmy', figsize=(9,6))
```

```python
df.plot.line(y=['FB','AAPL','MSFT'],color='bmy', figsize=(10,6))
```

```python
df.plot(y='FB', figsize=(10,6), title='Facebook Stock', ylabel='USD')
```

```python
df_3Months = df.resample(rule='M').mean()[-3:]
print(df_3Months)
```

```python
df_3Months.plot(kind='bar', figsize=(10,6), ylabel='Price')
```

```python
df_3Months.plot(kind='barh', figsize=(9,6))
```

```python
df_3Months.plot(kind='bar', stacked=True, figsize=(9,6))
```

3. Histogram: Plot histogram for all three stocks and change the interval to 15.

```python
df[['MSFT','FB','AAPL']].plot(kind='hist', bins=15, alpha=0.6, figsize=(9,6))
```

```python
df[['MSFT','FB','AAPL']].plot(kind='hist', bins=15, alpha=0.6, stacked=True, figsize=(9,6))
```

4. Box Plot: No Change

```python
df.plot(kind='box', figsize=(9,6))
```

```python
df.plot(kind='box', vert=False, figsize=(9,6))
```

```python
df.plot(kind='area', figsize=(9,6))
```

```python
df.plot(kind='area', stacked=False, figsize=(9,6))
```

```python
df_3Months.index=['March','April','May']
df_3Months.plot(kind='pie', y ='AAPL', legend=False, autopct='%.f')
```

```python
df_3Months.plot(kind='pie', legend=False, autopct='%.f', subplots=True, figsize=(14,8))
```

5. Scatter Plot: Use subplots to print the scatter plot for

        ‘MSFT’ vs ‘FB’ in blue color
